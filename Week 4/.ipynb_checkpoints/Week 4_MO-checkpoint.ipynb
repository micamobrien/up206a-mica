## Isochrone Mapping

I now want to try to create an isochrone map with the affordable housing dataset. I start by importing the osmnx library.

Now I want to bring in the street network map for the area within 20 minutes of the affordable housing development.

```python
address = 'South Menlo Avenue, Harbor Gateway, Los Angeles, Los Angeles County, California, 90247, United States'
network_type = 'walk'
trip_times = [5, 10, 15, 20] 
meters_per_minute = 75 
```

I download the network map and check the time it takes to pull the data. 

```python
%%time

G = ox.graph_from_address(address, network_type=network_type, dist = 2000)
```

I plot the data before working with the nodes and edges.

```python
fig, ax = ox.plot_graph(G,figsize=(12,12))
```

I convert the nodes and edges to geodataframes.

```python
gdf_nodes, gdf_edges = ox.graph_to_gdfs(G)
```

I take a look at the nodes dataframe using the .sample() command. It looks like the x and y columns are the lat and long coordinates from the geometrty column.

```python
gdf_nodes.sample(5)
```

Then I look at the edges dataframe using the .sample() command. The geometry column is now populated with 'linestring'.

```python
gdf_edges.sample(5)
```

I create a new column called 'walk_time', which takes the 'length' variable from the edges dataframe divided by the 'meters_per_minute' variable I created above. I use the .sample() command to make sure the new column is showing up. 

```python
gdf_edges['walk_time'] = gdf_edges['length']/meters_per_minute
```

```python
gdf_edges.sample(5)
```

I assign the 4 trip times with unique colors and print out the hex codes. 

```python
iso_colors = ox.plot.get_colors(n=len(trip_times), 
                                cmap='plasma', 
                                start=0, 
                                return_hex=True)
print(trip_times)
print(iso_colors)
```

I zip the times and colors as 'time_color' to use later. 

```python
time_color = list(zip(trip_times, iso_colors))
time_color
```

I'm going to identify the center node, which I believe should be the address of the affordable housing address I identified above. I first identify the boundary box coordinates. 

```python
minx, miny, maxx, maxy = gdf_nodes.geometry.total_bounds
```

I now calculate the centroid. 

```python
centroid_x = (maxx-minx)/2 + minx
centroid_y = (maxy-miny)/2 + miny
```

I identify the nearest node to the center point and pull out the record. 

```python
center_node = ox.distance.nearest_nodes(G,Y=centroid_y,X=centroid_x)
```

```python
gdf_nodes.loc[[center_node]]
```

I pull out the number of nodes within 5, 10, 15, and 20 minutes from the center node and print them out as strings. 

```python
for time, color in list(time_color):

    # for each trip time, create an egograph of nodes that fall within that distance
    subgraph = nx.ego_graph(G, center_node, radius=time)

    print('There are ' + str(len(subgraph.nodes())) + ' nodes within ' + str(time) + ' minutes ')
    
    # for each of those nodes, update the gdf_nodes dataframe and assign it with its associated distance color
    for node in subgraph.nodes():
        gdf_nodes.loc[node,'time'] = str(time) + ' mins'
        gdf_nodes.loc[node,'color'] = color
```

I assign the NA variables to a color so they don't get incorrectly grouped.

```python
gdf_nodes['color'].fillna('#cccccc', inplace=True)
```

I use the .sample command to make sure that the #cccccc color was correctly applied to the NAs. Looks like it worked.

```python
gdf_nodes.sample(5)
```

I want to pull all of the nodes that fall within 5 minutes walking distance from the center point. It looks like no records fall into this category, which is odd because above it says that there are 74 nodes that are within 5 minutes.

```python
gdf_nodes[gdf_nodes['time']=='5 mins']
```

The same issue arises when I try 10 and 15 minutes, which is strange. The only time that seems to work is 20 minutes, which I will use moving forward. I save the dataframe as "gdf_20".

```python
gdf_20 = gdf_nodes[gdf_nodes['time']=='20 mins']
```

I plot gdf_20.

```python
gdf_20.plot(figsize=(10,10),color=gdf_20.color)
```

When I plot all of the nodes, it still looks like only the nodes within 20 minutes are showing up. I'm not quite sure what to make of this.

```python
gdf_nodes.plot(figsize=(10,10),
               color=gdf_nodes['color'],
               markersize=3)
```

I create a convex hull polygon to connect all of the points.

```python
gdf_20.unary_union.convex_hull
```

Now I can dissolve the nodes by time. Only 20 minutes shows up again.

```python
isochrones = gdf_nodes.dissolve("time")
isochrones
```

I create a new convex hull for the 20 minutes row.

```python
isochrones = isochrones.convex_hull.reset_index()
isochrones
```

I rename the columns as 'time' and 'geometry'. 

```python
isochrones.columns=['time','geometry']
```

I plot the isochrone.

```python
isochrones.plot(figsize=(10,10),alpha=0.2,cmap='plasma')
```

I end by situating the isochrone on a base map. This didn't turn out exactly the way I wanted - the isochrone only covers 20 minutes walking distance from the centroid. I'm still not quite sure why the 5, 10, and 15 minute distances don't show up and hope to do some investigating into this next week. 

```python
fig, ax = plt.subplots(figsize=(10,15))

# add the isochrones
isochrones.plot(alpha=0.4, 
                ax=ax, 
                column='time', 
                cmap='plasma', 
                legend=True,
                zorder=20)

# add the center node in red
gdf_nodes.loc[[center_node]].plot(ax=ax,color='r', zorder=30)

# add all nodes
gdf_nodes.plot(ax=ax, 
               markersize=1, 
               zorder=10)

# add the edges
gdf_edges.plot(ax=ax,
               linewidth=0.5,
               alpha=0.2,
               zorder=10)

# hide the axis
ax.axis('off')

# give it a title
ax.set_title('Walking areas from the center of ' + address,fontsize=16,pad=10)

# add the basemap
ctx.add_basemap(ax,source=ctx.providers.CartoDB.Positron)
```
